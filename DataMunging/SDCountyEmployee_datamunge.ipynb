```python
import pandas as pd
import numpy as np
import csv
```

```python
# had to resave because previous table was not encoded well
filepath = './datasets/test.csv'
```

```python
df = pd.read_csv(filepath, memory_map=True, engine='python', quoting=csv.QUOTE_NONE)
```

```python
df
```

```python
# remove unneccessary quotes so we can actually use in jupyter
df['Position'] = df['Position'].str.replace('"', '')
```

```python
# Remove Employer Type because everything is city
df = df.drop(columns=['EmployerType'])
```

```python
# Remove all '\xa0' so we can import to jupyter (this was a pain in the ass )
df = df.replace('\xa0', '', regex=True)
```

```python
# Replace False with NaN
df.replace(False, np.nan, inplace=True)
df.replace("FALSE", np.nan, inplace=True)
```

```python
# Check df to see if there are NaN
cols = df.columns

for i in cols:
    if df[i].isna().sum() > 0:
        print(i, " contains NaN") 
```

```python
# we are interested in grouping by ["police", "fire", "city attorney", "sheriff", "teacher", "educator", "school"]
df["DepartmentOrSubdivision"].unique()
```

```python
# Strip leading and trailing whitespace
df["DepartmentOrSubdivision"] = df["DepartmentOrSubdivision"].str.strip()
```

```python
# check nan departments
df["DepartmentOrSubdivision"].isna().value_counts()
```

```python
# remove nan because they look pretty useless ngl
df = df.dropna(subset=["DepartmentOrSubdivision"])
```

```python
# Check df to see if there are NaN
cols = df.columns

for i in cols:
    if df[i].isna().sum() > 0:
        print(i, " contains NaN with ", df[df[i].isna()].shape[0], "rows") 
        display(df[df[i].isna()])
```

```python
def symmetric_difference(list1, list2):
    """
    Compute the symmetric difference between two lists.
    
    Parameters:
    - list1 (list): The first list.
    - list2 (list): The second list.
    
    Returns:
    - list: The symmetric difference between list1 and list2.
    """
    set1 = set(list1)
    set2 = set(list2)
    
    # Compute symmetric difference using sets
    symmetric_diff_set = set1.symmetric_difference(set2)
    
    # Convert set back to list
    symmetric_diff_list = list(symmetric_diff_set)
    
    return symmetric_diff_list
```

```python
# Group by df["DepartmentOrSubdivision"] into keywords : ["police", "fire", "city attorney", "sheriff", "teacher", "educator", "school"]

depts = list(df["DepartmentOrSubdivision"].unique())

police_depts = [i for i in depts if "pol" in i.lower()]
fire_depts = [i for i in depts if "fire" in i.lower() or "sdfd" in i.lower()]
attorney_depts = [i for i in depts if "attorney" in i.lower()]
sheriff_depts = [i for i in depts if "sheriff" in i.lower()]
teacher_depts = [i for i in depts if "teacher" in i.lower() or "school" in i.lower()]
giga_depts = police_depts + fire_depts + sheriff_depts + teacher_depts + attorney_depts

other_depts = symmetric_difference(depts, giga_depts)
```

```python
police_df = df[df["DepartmentOrSubdivision"].isin(police_depts)].reset_index()
fire_df = df[df["DepartmentOrSubdivision"].isin(fire_depts)].reset_index()
attorney_df = df[df["DepartmentOrSubdivision"].isin(attorney_depts)].reset_index()
sheriff_df = df[df["DepartmentOrSubdivision"].isin(sheriff_depts)].reset_index()
teacher_df = df[df["DepartmentOrSubdivision"].isin(teacher_depts)].reset_index()
other_df = df[df["DepartmentOrSubdivision"].isin(other_depts)].reset_index()
```
